```python
import pandas as pd
```

```python
year = 2020
month = '03'
ending = 'xlsx'
if year < 2019 and int(month) < 12:
    ending = 'xls'    
data_path = 'https://www.kba.de/SharedDocs/Publikationen/DE/Statistik/Fahrzeuge/FZ/{y}_monatlich/FZ10/fz10_{y}_{m}_{e}.{e}?__blob=publicationFile&v=2'.format(y=year,m=month,e=ending)
print(data_path)
df = pd.read_excel(data_path,sheet_name=-1,header=[7,8]).drop(df.columns[0],axis=1)
```
